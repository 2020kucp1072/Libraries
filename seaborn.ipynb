```python
'''
    @Author: VEMULA DILEEP
    @Date: 25-10-24
    @Last Modified by: VEMULA DILEEP
    @Last Modified time: 17:55
    @Title : Python programs for seaborn
'''


import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

def bar_plot_sex_survived():
    """
    Description:
        This function loads the Titanic dataset and creates a bar plot of 'sex' against 'survived'.
    Parameters:
        None
    Return:
        None
    """
    url = "https://github.com/mwaskom/seaborn-data/raw/master/titanic.csv"
    data = pd.read_csv(url)
    survived_data = data.groupby('sex')['survived'].sum()
    
    sns.barplot(x=survived_data.index, y=survived_data.values)
    plt.title("Survival Rate by Sex")
    plt.xlabel("Sex")
    plt.ylabel("Number of Survivors")
    plt.show()

bar_plot_sex_survived()
```

```python
def point_plot_sex_survived():
    """
    Description:
        This function loads the Titanic dataset and creates a point plot of 'sex' against 'survived'.
    Parameters:
        None
    Return:
        None
    """
    url = "https://github.com/mwaskom/seaborn-data/raw/master/titanic.csv"
    data = pd.read_csv(url)
    survived_data = data.groupby('sex')['survived'].sum()
    sns.pointplot(x=survived_data.index, y=survived_data.values)
    plt.title("Survival Rate by Sex (Point Plot)")
    plt.show()

point_plot_sex_survived()
```

```python
def scatter_plot_day_total_bill():
    """
    Description:
        This function loads the Tips dataset and creates a scatter plot of 'day' against 'total_bill'.
    Parameters:
        None
    Return:
        None
    """
    url = "https://raw.githubusercontent.com/mwaskom/seaborn-data/master/tips.csv"
    data = pd.read_csv(url)
    sns.scatterplot(x="day", y="total_bill", data=data)
    plt.title("Total Bill by Day (Scatter Plot)")
    plt.show()

scatter_plot_day_total_bill()
```

```python
def violin_plot_sex_total_bill():
    """
    Description:
        This function loads the Tips dataset and creates a violin plot of 'sex' against 'total_bill'.
    Parameters:
        None
    Return:
        None
    """
    url = "https://raw.githubusercontent.com/mwaskom/seaborn-data/master/tips.csv"
    data = pd.read_csv(url)
    sns.violinplot(x="sex", y="total_bill", data=data)
    plt.title("Total Bill by Sex (Violin Plot)")
    plt.show()

violin_plot_sex_total_bill()
```

```python
def violin_plot_species_sepal_length():
    """
    Description:
        This function loads the Iris dataset and creates a violin plot of 'species' against 'sepal_length'.
    Parameters:
        None
    Return:
        None
    """
    url = "https://github.com/mwaskom/seaborn-data/raw/master/iris.csv"
    data = pd.read_csv(url)
    sns.violinplot(x="species", y="sepal_length", data=data)
    plt.title("Sepal Length by Species (Violin Plot)")
    plt.show()

violin_plot_species_sepal_length()
```

```python
def box_plot_lifeExp_continent():
    """
    Description:
        This function loads the Gapminder dataset and creates a box plot of 'lifeExp' by 'continent'.
    Parameters:
        None
    Return:
        None
    """
    url = "https://raw.githubusercontent.com/resbaz/r-novice-gapminder-files/master/data/gapminder-FiveYearData.csv"
    data = pd.read_csv(url)
    sns.boxplot(x="continent", y="lifeExp",data=data,palette='coolwarm')
    plt.title("Life Expectancy by Continent (Box Plot)")
    plt.show()

box_plot_lifeExp_continent()
```

```python
def box_plot_day_tip():
    """
    Description:
        This function loads the Tips dataset and creates a box plot of 'day' by 'tip'.
    Parameters:
        None
    Return:
        None
    """
    url = "https://raw.githubusercontent.com/mwaskom/seaborn-data/master/tips.csv"
    data = pd.read_csv(url)
    sns.boxplot(x="day", y="tip", data=data,palette='coolwarm')
    plt.title("Tip by Day (Box Plot)")
    plt.show()

box_plot_day_tip()
```

```python
def swarm_plot_total_bill_size():
    """
    Description:
        This function loads the Tips dataset and creates a swarm plot of 'total_bill' against 'size'.
    Parameters:
        None
    Return:
        None
    """
    url = "https://raw.githubusercontent.com/mwaskom/seaborn-data/master/tips.csv"
    data = pd.read_csv(url)
    sns.swarmplot(x="size", y="total_bill", data=data,size=4,palette='coolwarm')
    plt.title("Total Bill by Size (Swarm Plot)")
    plt.show()

swarm_plot_total_bill_size()
```

```python
def swarm_plot_total_bill_day():
    """
    Description:
        This function loads the Tips dataset and creates a swarm plot of 'total_bill' against 'day'.
    Parameters:
        None
    Return:
        None
    """
    url = "https://raw.githubusercontent.com/mwaskom/seaborn-data/master/tips.csv"
    data = pd.read_csv(url)
    sns.swarmplot(x="day", y="total_bill", data=data,size=4)
    plt.title("Total Bill by Day (Swarm Plot)")
    plt.show()

swarm_plot_total_bill_day()
```

```python
def main():
    
    bar_plot_sex_survived()
    point_plot_sex_survived()
    scatter_plot_day_total_bill()
    violin_plot_sex_total_bill()
    violin_plot_species_sepal_length()
    box_plot_lifeExp_continent()
    box_plot_day_tip()
    swarm_plot_total_bill_size()
    swarm_plot_total_bill_day()
    
if __name__=="__main__":
    main()
    
```
