```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

def bar_plot_sex_survived():
    """
    Description:
        This function loads the Titanic dataset and creates a bar plot of 'sex' against 'survived'.
    Parameters:
        None
    Return:
        None
    """
    url = "https://github.com/mwaskom/seaborn-data/raw/master/titanic.csv"
    data = pd.read_csv(url)
    survived_data = data.groupby('sex')['survived'].sum()
    
    sns.barplot(x=survived_data.index, y=survived_data.values)
    plt.title("Survival Rate by Sex")
    plt.xlabel("Sex")
    plt.ylabel("Number of Survivors")
    plt.show()

bar_plot_sex_survived()
```

```python
def main():
    bar_plot_sex_survived()

if __name__=="__main__":
    main()
```
