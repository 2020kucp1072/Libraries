```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

def bar_plot_sex_survived():
    """
    Description:
        This function loads the Titanic dataset and creates a bar plot of 'sex' against 'survived'.
    Parameters:
        None
    Return:
        None
    """
    url = "https://github.com/mwaskom/seaborn-data/raw/master/titanic.csv"
    data = pd.read_csv(url)
    survived_data = data.groupby('sex')['survived'].sum()
    
    sns.barplot(x=survived_data.index, y=survived_data.values)
    plt.title("Survival Rate by Sex")
    plt.xlabel("Sex")
    plt.ylabel("Number of Survivors")
    plt.show()

bar_plot_sex_survived()
```

```python
def point_plot_sex_survived():
    """
    Description:
        This function loads the Titanic dataset and creates a point plot of 'sex' against 'survived'.
    Parameters:
        None
    Return:
        None
    """
    url = "https://github.com/mwaskom/seaborn-data/raw/master/titanic.csv"
    data = pd.read_csv(url)
    survived_data = data.groupby('sex')['survived'].sum()
    sns.pointplot(x=survived_data.index, y=survived_data.values)
    plt.title("Survival Rate by Sex (Point Plot)")
    plt.show()

point_plot_sex_survived()
```

```python
def scatter_plot_day_total_bill():
    """
    Description:
        This function loads the Tips dataset and creates a scatter plot of 'day' against 'total_bill'.
    Parameters:
        None
    Return:
        None
    """
    url = "https://raw.githubusercontent.com/mwaskom/seaborn-data/master/tips.csv"
    data = pd.read_csv(url)
    sns.scatterplot(x="day", y="total_bill", data=data)
    plt.title("Total Bill by Day (Scatter Plot)")
    plt.show()

scatter_plot_day_total_bill()
```

```python
def violin_plot_sex_total_bill():
    """
    Description:
        This function loads the Tips dataset and creates a violin plot of 'sex' against 'total_bill'.
    Parameters:
        None
    Return:
        None
    """
    url = "https://raw.githubusercontent.com/mwaskom/seaborn-data/master/tips.csv"
    data = pd.read_csv(url)
    sns.violinplot(x="sex", y="total_bill", data=data)
    plt.title("Total Bill by Sex (Violin Plot)")
    plt.show()

violin_plot_sex_total_bill()
```

```python
def main():
    bar_plot_sex_survived()
    point_plot_sex_survived()
    scatter_plot_day_total_bill()
    violin_plot_sex_total_bill()
    
if __name__=="__main__":
    main()
    
```
