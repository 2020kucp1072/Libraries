```python


'''
    @Author: VEMULA DILEEP
    @Date: 25-10-24
    @Last Modified by: VEMULA DILEEP
    @Last Modified time: 17:55
    @Title : Python programs for matplotlib - line plot
'''
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# 1. Draw a line with labels and a title
def draw_basic_line():
    """Draws a basic line with labeled axes and a title.
    
    Parameters: None
    Returns: None
    """
    x = [1, 2, 3, 4, 5]
    y = [2, 3, 5, 7, 11]
    plt.plot(x, y)
    plt.xlabel('X axis label')
    plt.ylabel('Y axis label')
    plt.title('Basic Line Plot')
    plt.show()


# 2. Draw a line using given axis values
def draw_line_with_values(x_values, y_values):
    """Draws a line plot for given x and y values.
    
    Parameters:
        x_values (list): List of x-axis values
        y_values (list): List of y-axis values
    Returns: None
    """
    plt.plot(x_values, y_values)
    plt.xlabel('X axis')
    plt.ylabel('Y axis')
    plt.title('Line Plot with Given Values')
    plt.show()


# 3. Draw a line using values from a text file
def draw_line_from_file(file_path):
    """Reads x and y values from a file and draws a line plot.
    
    Parameters:
        file_path (str): Path to the text file with x and y values
    Returns: None
    """
    data = np.loadtxt(file_path)
    x_values, y_values = data[:, 0], data[:, 1]
    plt.plot(x_values, y_values)
    plt.xlabel('X axis')
    plt.ylabel('Y axis')
    plt.title('Line Plot from File Data')
    plt.show()


# 4. Draw a line chart for Alphabet Inc. financial data
def draw_financial_line_chart(file_path):
    """Draws a line chart for Alphabet Inc. data.
    
    Parameters:
        file_path (str): Path to the financial data CSV file
    Returns: None
    """
    data = pd.read_csv(file_path, parse_dates=['Date'])
    plt.plot(data['Date'], data['Close'], label='Close')
    plt.xlabel('Date')
    plt.ylabel('Close Price')
    plt.title('Alphabet Inc. Close Prices')
    plt.show()


# 5. Plot multiple lines on the same plot with legends
def plot_multiple_lines():
    """Plots multiple lines with legends on the same plot.
    
    Parameters: None
    Returns: None
    """
    x = [1, 2, 3, 4, 5]
    plt.plot(x, [1, 2, 3, 4, 5], label="Line 1")
    plt.plot(x, [2, 3, 4, 5, 6], label="Line 2")
    plt.legend()
    plt.show()


# 6. Plot lines with different widths and colors
def plot_lines_with_styles():
    """Plots lines with varying styles, widths, and colors.
    
    Parameters: None
    Returns: None
    """
    x = [1, 2, 3, 4, 5]
    plt.plot(x, [1, 2, 3, 4, 5], label="Line 1", color="blue", linewidth=2)
    plt.plot(x, [2, 3, 4, 5, 6], label="Line 2", color="red", linewidth=4)
    plt.legend()
    plt.show()


# 7. Plot lines with different styles
def plot_lines_with_different_styles():
    """Plots multiple lines with different line styles.
    
    Parameters: None
    Returns: None
    """
    x = [1, 2, 3, 4, 5]
    plt.plot(x, [1, 2, 3, 4, 5], 'r-', label="Solid Line")
    plt.plot(x, [2, 3, 4, 5, 6], 'b--', label="Dashed Line")
    plt.legend()
    plt.show()


# 8. Plot lines with markers
def plot_lines_with_markers():
    """Plots lines with markers at each data point.
    
    Parameters: None
    Returns: None
    """
    x = [1, 2, 3, 4, 5]
    plt.plot(x, [1, 2, 3, 4, 5], 'ro-', label="Line with Circle Markers")
    plt.plot(x, [2, 3, 4, 5, 6], 'bs--', label="Line with Square Markers")
    plt.legend()
    plt.show()


# 9. Display and set new axis limits
def display_and_set_axis_limits():
    """Displays and sets custom axis limits for the current plot.
    
    Parameters: None
    Returns: None
    """
    x = [1, 2, 3, 4, 5]
    y = [2, 4, 6, 8, 10]
    plt.plot(x, y)
    print("Current axis limits:", plt.axis())
    plt.axis([0, 6, 0, 12])
    plt.show()

    # 10. Plot quantities with x and y positions
def plot_quantities(x, y):
    """Plots quantities using given x and y positions.
    
    Parameters:
        x (list): List of x-axis positions
        y (list): List of y-axis positions
    Returns: None
    """
    plt.plot(x, y, 'bo-')
    plt.xlabel('X positions')
    plt.ylabel('Y quantities')
    plt.title('Quantities at Given Positions')
    plt.show()

# 11. Plot multiple lines with different format styles in one command using arrays
def plot_multiple_lines_with_styles():
    """Plots several lines with different format styles in one command.
    
    Parameters: None
    Returns: None
    """
    x = np.array([1, 2, 3, 4, 5])
    y1 = x
    y2 = x**2
    y3 = x**3
    plt.plot(x, y1, 'g-', x, y2, 'b--', x, y3, 'r-.')
    plt.show()

# 12. Create multiple types of charts
def create_multiple_charts():
    """Creates multiple types of charts: line, scatter, and bar.
    
    Parameters: None
    Returns: None
    """
    x = np.array([1, 2, 3, 4, 5])
    y = np.array([2, 4, 6, 8, 10])

    plt.figure(figsize=(10, 5))

    # Line plot
    plt.subplot(1, 3, 1)
    plt.plot(x, y, 'r-')
    plt.title('Line Plot')

    # Scatter plot
    plt.subplot(1, 3, 2)
    plt.scatter(x, y)
    plt.title('Scatter Plot')

    # Bar plot
    plt.subplot(1, 3, 3)
    plt.bar(x, y)
    plt.title('Bar Plot')

    plt.tight_layout()
    plt.show()

# 13. Line chart of Alphabet Inc. data with customized grid lines
def draw_financial_chart_with_custom_grid(file_path):
    """Draws a line chart with a custom grid style.
    
    Parameters:
        file_path (str): Path to the financial data CSV file
    Returns: None
    """
    data = pd.read_csv(file_path, parse_dates=['Date'])
    plt.plot(data['Date'], data['Close'], label='Close')
    plt.grid(linestyle='-', linewidth=0.5, color='blue')
    plt.xlabel('Date')
    plt.ylabel('Close Price')
    plt.title('Alphabet Inc. Close Prices with Custom Grid')
    plt.show()

# 14. Line chart of Alphabet Inc. data with major and minor grids, no ticks
def draw_chart_with_major_minor_grids(file_path):
    """Draws a line chart with major and minor grid lines, without ticks.
    
    Parameters:
        file_path (str): Path to the financial data CSV file
    Returns: None
    """
    data = pd.read_csv(file_path, parse_dates=['Date'])
    plt.plot(data['Date'], data['Close'], label='Close')
    plt.grid(which='major', linestyle='-', linewidth=0.5, color='blue')
    plt.minorticks_on()
    plt.grid(which='minor', linestyle=':', linewidth=0.5, color='gray')
    plt.tick_params(left=False, bottom=False)
    plt.xlabel('Date')
    plt.ylabel('Close Price')
    plt.title('Alphabet Inc. Prices with Major & Minor Grids, No Ticks')
    plt.show()

# 15. Create multiple subplots in one figure
def create_multiple_subplots():
    """Creates multiple subplots in a single figure.
    
    Parameters: None
    Returns: None
    """
    x = np.array([1, 2, 3, 4, 5])
    y1 = x
    y2 = x**2
    y3 = x**3

    plt.figure(figsize=(10, 6))

    # First subplot
    plt.subplot(3, 1, 1)
    plt.plot(x, y1, 'r-')
    plt.title('y = x')

    # Second subplot
    plt.subplot(3, 1, 2)
    plt.plot(x, y2, 'b--')
    plt.title('y = x^2')

    # Third subplot
    plt.subplot(3, 1, 3)
    plt.plot(x, y3, 'g-.')
    plt.title('y = x^3')

    plt.tight_layout()
    plt.show()

# Update main function to include all 15 functions
def main():
    draw_basic_line()
    draw_line_with_values([1, 2, 3, 4, 5], [2, 4, 1, 3, 5])
    draw_line_from_file('test.txt')  # Ensure 'test.txt' file exists
    draw_financial_line_chart('fdata.csv')  # Ensure 'fdata.csv' file exists
    plot_multiple_lines()
    plot_lines_with_styles()
    plot_lines_with_different_styles()
    plot_lines_with_markers()
    display_and_set_axis_limits()
    plot_quantities([1, 2, 3, 4, 5], [2, 3, 5, 7, 11])
    plot_multiple_lines_with_styles()
    create_multiple_charts()
    draw_financial_chart_with_custom_grid('fdata.csv')
    draw_chart_with_major_minor_grids('fdata.csv')
    create_multiple_subplots()

# Run the main function
if __name__ == "__main__":
    main()
```
